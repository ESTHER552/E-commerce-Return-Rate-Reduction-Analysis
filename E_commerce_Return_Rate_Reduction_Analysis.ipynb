```python
import pandas as pd

df = pd.read_csv('ecommercedataset.csv')
```

```python
df.duplicated()
```

```python
##Change date format
df['Order_Date'] = pd.to_datetime(df['Order_Date'], dayfirst=True, errors='coerce')
df['Order_Date'] = df['Order_Date'].dt.strftime('%d/%m/%Y')

df['Return_Date'] = pd.to_datetime(df['Return_Date'], dayfirst=True, errors='coerce')
df['Return_Date'] = df['Return_Date'].dt.strftime('%d/%m/%Y')
```

```python
##Replace empty Return date with null
df.fillna({'Return_Date': 'Null'}, inplace=True)
df.fillna({'Return_Season': 'NUll'}, inplace=True)
df.fillna({'Days_to_return': 'Null'}, inplace=True)
```

```python
##Re-calculating days_to_return
df['Order_Date'] = pd.to_datetime(df['Order_Date'], dayfirst=True, errors='coerce')
df['Return_Date'] = pd.to_datetime(df['Return_Date'], dayfirst=True, errors='coerce')
df['Days_to_Return'] = (df['Return_Date'] - df['Order_Date']).dt.days
```

```python
##Remove white spaces
df = df.apply(lambda x: x.str.replace(" ", "") if x.dtype == "object" else x)
```

```python
##Validate Date Logic
df[df['Return_Date'] < df['Order_Date']]
```

```python
mask2 = (df['Return_Date'] < df['Order_Date']) & (df['Days_to_Return'] <= 0)

df.loc[mask2, ['Order_Date', 'Return_Date']] = \
df.loc[mask2, ['Return_Date', 'Order_Date']].values
```

```python
# Make sure dates are datetime
df['Order_Date'] = pd.to_datetime(df['Order_Date'], dayfirst=True, errors='coerce')
df['Return_Date'] = pd.to_datetime(df['Return_Date'], dayfirst=True, errors='coerce')

# Recalculate Days_to_Return for only returned items
df['Days_to_Return'] = (df['Return_Date'] - df['Order_Date']).dt.days
```
